# Excercise 1

```python
import matplotlib.pyplot as plt
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='red', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

# Excercise 2

```python
import numpy as np
import matplotlib.pyplot as plt
arrayx=np.array([i for i in range(0,501,1)])/10
arrayy=np.exp(arrayx*(-1/10))*np.sin(arrayx)
plt.plot(arrayx, arrayy)
plt.title(r'$e^{-x/10}sin(x)$',fontsize=20)
plt.xlabel('x',color='darkorange',fontsize=15)
plt.ylabel('y',color='darkorange',fontsize=15,rotation='horizontal')
plt.grid(alpha=.25)
plt.ylim(-1,1)
plt.show()
```
